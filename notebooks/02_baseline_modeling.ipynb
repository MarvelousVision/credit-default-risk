## DummyClassifier Baseline
Precision, recall, and F1 are 0 because the dummy model never predicts default.

The DummyClassifier predicts only the majority class, Fully Paid. It achieves 74.5% accuracy on validation, but recall, precision, and F1 for default detection are 0. The confusion matrix shows that all 54,659 default cases were missed. This confirms that accuracy alone is not suitable for this project. Future models must improve recall, PR-AUC, and business cost compared with this no-skill baseline.

| Balanced Logistic Regression |   0.6560 |    0.3744 | 0.5227 | 0.4363 |  0.6651 | 0.3932 |


The DummyClassifier predicts only the majority class and misses all defaults. Normal Logistic Regression has high precision but very low recall, meaning it predicts default only rarely and misses most risky loans. Balanced Logistic Regression greatly improves recall and F1, detecting many more defaulted loans, but it also increases false positives.

For this credit-risk project, false negatives are more expensive because they represent risky borrowers being approved. Therefore, Balanced Logistic Regression is a stronger baseline than normal Logistic Regression, even though its accuracy is lower. The next step is to compare stronger models and later tune the decision threshold using a business cost metric.
