```python
import numpy as np 
def e_mod(X,y_proba, y_test, threshold, review_effectiveness):


    y_pred = (y_proba >= threshold).astype(int)

    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    fn_loan_amounts = X.loc[(y_test == 1) & (y_pred == 0), 'loan_amnt']
    fn_cost = fn_loan_amounts.sum() * 0.60
    fp_cost = fp * 100
    review_cost = (fp + tp) * 100
    total_cost = fn_cost + review_cost
    estimated_business_cost= fn_cost + fp_cost
    flagged_rate= np.mean(y_pred)
    f1 = f1_score(y_test, y_pred)
    tp_loan_amounts = X.loc[(y_test == 1) & (y_pred == 1), 'loan_amnt']
    tp_residual_loss = tp_loan_amounts.sum() * fn_cost * (1 - review_effectiveness)
    return {
        'threshold': threshold,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'fn_loss': fn_cost,
        'tp_residual_loss': tp_residual_loss,
        'review_cost': review_cost,
        'total_cost': total_cost,
        'flagged_rate': flagged_rate,
        'fn_count': fn,
        'fp_count': fp,
        'tp_count': tp,
        'tn_count': tn,
        'confusion_matrix': cm
    }
thresholds = np.arange(0.080, 0.101, 0.001)
y_proba = lmodel.predict_proba(X_val)[:,1]
for i in thresholds:
    print(i, e_mod(X_val,y_proba, y_val, i, 1.00))
```

|      0.15 |     0.351 |  0.702 |  71,125 | 16,267 | strong recall, fewer FP than 0.10            |
|      0.20 |     0.395 |  0.506 |  42,433 | 26,991 | balanced option                              |

The default threshold 0.5 is too conservative for this credit-risk problem. It gives high precision but very low recall, missing most defaulted loans. Lower thresholds substantially improve recall, but increase false positives. Thresholds between 0.15 and 0.25 are the most promising candidates. The final threshold should be selected using a business cost metric that makes false negatives more expensive than false positives.
